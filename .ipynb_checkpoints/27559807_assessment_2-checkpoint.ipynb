### Function to validate and format the date

```python
# Function to parse the start and end date 

def convert_date(s):
    if len(s) == 4:      # Take only values with year
        return s
    elif len(s) < 7:   # Take only values with month and year
        return s[: -4].zfill(2) + '-' + s[-4: ]
    elif len(s) >= 7:  # Take only values with day, month and year
        return s[: -6].zfill(2) + '-' + s[-6: -4].zfill(2) + '-' + s[-4:]
    else:
        return None

```

### Apply the 'convert_date' function

```python
df3.Start = df3.Start.apply(convert_date).astype(str)
```

```python
df3.End = df3.End.apply(convert_date).astype(str)
```

### Process to validate the date

```python
# Function to convert the string to date

def to_date(strdate,dtformat):
    st = datetime.datetime.strptime(strdate, dtformat).date()
    return st
```

```python
# Function to return the date format based on length of the string

def dt_format(v1):
    
    if v1==4:
        dateformat = "%Y"
    elif v1==7:
        dateformat = "%M-%Y"
    elif v1==10:
        dateformat = "%d-%m-%Y"
    else:    
        print 'Unrecognised'
    return  dateformat  
```

```python
# Process to identify the invalid and irregular dates

for i in df3.itertuples():
    try: 
        v1 = min(len(i[1]),len(i[2]))        
        st = to_date(i[1][-v1:], dt_format(v1))
        ed = to_date(i[2][-v1:], dt_format(v1))

        if st>ed:
            print "Integrity constraint violation "+str(i[0])+' '+str(st)+' '+str(ed)+' '+i[3]+' '+i[4]+' '+i[5]+' '+i[6]+' '+str(i[7])+' '+str(i[8])

    except (ValueError):
            string = 'Invalid Dates '+str(i[0])+' '+str(i[1])+' '+str(i[2])+' '+str(i[3])+' '+str(i[4])+' '+str(i[5])+' '+str(i[6])+' '+str(i[7])
            print string
            pass
```

### Data Imputation

#### Rules used to Fix Dates
**For Start Dates**
 - If the month is "00" set the month to "01"
 - If the day is "00" set the day to "01"
 - If the day is greater than the days in the month, set to the last day of the month

**For End Dates**
 - If the month is "00" set the month to "12"
 - If the day is "00" set to the last day of the month
 - If the day is greater than the days in the month, set to the last day of the month

### Fixing Date Format

```python
# Dictionary that specifies the number of days in each month
monthDict = {"01":"31", "02":"28", "03":"31", "04":"30", "05":"31", "06":"30",
             "07":"31", "08":"31", "09":"30", "10":"31", "11":"30", "12":"31"}
```

```python
# Function to convert dates to a valid date in ddmmyyyy format
def fixDate(value, useMax=False):
    if len(value) < 10:
        value = to_date(value,dt_format(len(value))) 
        value = value.strftime("%d-%m-%Y")

    day = value[0:2]                                  # Extract the day component 
    month = value[3:5]                                # Extract the month component
    year = value[6:10]                                 # Extract the year component    

    if month == "00":                                 # If the month is "00" ... 
        month = "12" if useMax else "01"              # Correct the month dependig on the the useMax flag
        value = day +'-'+ month+'-'+year              # Date value to return

    if day == "00":                                   # If the day is "00" ... 
        day = monthDict[month] if useMax else "01"    # Correct the day depending on the useMax flag
        value = day +'-'+ month+'-'+year                    # Date value to return

    try:
        dt.datetime.strptime(value, "%d-%m-%Y")         # Check the date is valid
    except:                                           # If not, the day must still be invalid
        day = monthDict[month]                        # Set to the last day of the month
        value = day +'-'+ month+'-'+year                    # Date value to return
        datetime.datetime.strptime(value, "%d-%m-%Y").date()      # Verify the date is now valid

    return value                                 # Return the fixed date

```

#### Set Datatype for Start to Datetime

```python
# Fix invalid dates in the Start column
df3["Start"] = df3.Start.apply(fixDate)
# Convert the Start column to datetime format
df3["Start"] = pd.to_datetime(df3["Start"], format="%d-%m-%Y")
```

#### Set Datatype for End to Datetime

```python
# Fix invalid dates in the End column
df3["End"] = df3.End.apply(fixDate, useMax=True)
# Convert the End column to datetime format
df3["End"] = pd.to_datetime(df3["End"], format="%d-%m-%Y")
```

### Defining Country Lookup List

```python
# Defining lookup values for country column

country_lookup={
"Bosnia And Herzegovina":["Bosnia-Hercegovenia"],
"Spain":["Canary Is"],
"Cape Verde":["Cape Verde Is Is"],
"Cook Islands":["Cook Is"],
"China":["Chine","China P Rep","Hong Kong (China)"],
"Dominican Republic":["Dominican Rep"],
"Germany":["Germany Fed Rep","Germany Dem Rep"],
"Gambia":["Gambia The"],
"Iran":["Iran Islam Rep"],
"Korea":["Korea Dem P Rep","Korea Rep"],
"Laos":["Lao P Dem Rep"],
"Libya":["Libyan Arab Jamah"],
"Macedonia":["Macedonia FRY"],
"Marshall Islands":["Marshall Is"],
"Micronesia":["Micronesia Fed States"],
"Moldova":["Moldova Rep"],
"Northern Mariana Islands":["Northern Mariana Is"],
"Palestine":["Palestine (West Bank)"],
"Sao Tome":["Sao Tome et Principe"],
"Solomon Islands":["Solomon Is"],
"Syrian Arab Republic":["Syrian Arab Rep"],
"Taiwan":["Taiwan (China)"],
"Tanzania":["Tanzania Uni Rep"],
"Turks and Caicos Islands":["Turks and Caicos Is"],
"United States":["USA","United Stats"],
"Virgin Islands U.S":["Virgin Is (US)"],
"Virgin Islands U.K" :["Virgin Is (UK)"],   
"Wallis and Futuna":["Wallis","Wallis and Futuna Is"],
"Yemen":["Yemen Arab Rep","Yemen P Dem Rep"],
"Zaire":["Zaire/Congo Dem Rep"]
    }

# Function to translate the error values into valid country names

def translate(in_value):
    out_value = in_value
    for k, v in country_lookup.iteritems():
        if out_value in v:
            out_value = k
    return out_value

```

```python
# apply translate function to all rows in the country column

df3['Country'] = df3['Country'].apply(translate)
```

- Using **country list**,we identified the inconsistent and lexical error values
- Working with those errors was easy by defining a **lookup table**

### Hampel Identifier

```python
#converting the dataframe to array to perform the MAD function below.
killedarray=tsk2.Killed.values
#finding the median of the killed column
median=tsk2.Killed.median()
median
```

```python
from math import fabs #to find absolute values
killedcolumn=[]
for a in killedarray:
    mad = fabs(a - median) #applying the formula for MAD
    killedcolumn.append(mad)
cleanedkilledcolumn=pd.DataFrame(killedcolumn)
```

```python
#Droppping and adding the (Median Absolute Deviation)-applied 'Killed' column to the dataframe 
tsk2=tsk2.reset_index(drop=True)
tsk2.Killed=cleanedkilledcolumn.copy()
```

- The flood which happened in Natal Province in South Africa has a estimated death toll over 400

- From the above the esitmated killed values are close enough with values referred in online
- It is genuine value
